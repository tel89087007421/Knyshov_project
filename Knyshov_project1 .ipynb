```python
def total_sales_group(total_sales):

    if total_sales <= 0.05:
        return '0.0-0.05'
    if total_sales <= 0.1:
        return '0.05-0.1'
    if total_sales <= 1:
        return '0.1-1'
    return '1-100'
```

```python
def popular_cinema_movies(row):

    critic_score = row['critic_score']
    total_sales_group = row['total_sales_group']

    if critic_score >= 70:
        if total_sales_group == '1-100':
            return 'popular'
#        if total_sales_group == '1-10':
#            return 'popular'

        return '-'

    return '-'
```
